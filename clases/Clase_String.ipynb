```python
s = "Avenida Vicuna Mackenna 4860"
#letras, digitos y espacios

c_letras = 0
c_digitos = 0
c_espacios = 0

for i in range(len(s)):
  c = s[i]
  if c == " ":
    c_espacios +=1
  elif c in "9876543210":
    c_digitos +=1
  else:
    c_letras +=1

print("Letras:",c_letras)
print("Digitos:",c_digitos)
print("Espacios:",c_espacios)
```
